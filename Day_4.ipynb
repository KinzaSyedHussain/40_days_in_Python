```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dateset
data= sns.load_dataset("titanic")
data
```

```python
# draw a line plot
sns.lineplot(x='age',y='fare',data=data)
plt.title('Line Plot')
plt.show()
```

```python
#draw a barplot
sns.barplot(x='who',y='alone', hue = "sex", data=data)
plt.title('Bar Plot')
plt.show()
```

```python
#order
sns.barplot(x='who',y='alone', hue = "sex", data=data, order=['child','woman','man'])
plt.title('Bar Plot')
plt.show()
```

```python
# changing the color
sns.barplot(x='who',y='alone', hue = "sex", data=data, order=['child','woman','man'], color = 'red')
plt.title('Bar Plot')
plt.show()

sns.barplot(x='who',y='alone', hue = "sex", data=data, order=['child','woman','man'], color = 'pink')
plt.title('Bar Plot')
plt.show()
```

```python
#using colour pallete of seabor
sns.barplot(x='who',y='alone', hue = "sex", data=data, order=['child','woman','man'], ci = None, palette='bright' )
plt.title('Bar Plot')
plt.show()

sns.barplot(x='who',y='alone', hue = "sex", data=data, order=['child','woman','man'], ci = None, palette='pastel' )
plt.title('Bar Plot')
plt.show()
```

```python
#horizontal plot
sns.barplot(x='fare',y='class', hue = "sex", data=data, estimator=numpy.mean, saturation=1)
plt.title('Bar Plot')
plt.show()
```

```python
sns.barplot(x='class',y='fare', hue = "sex", data=data, linewidth=2.5, facecolor=(1, 1,  1, 0),
            errcolor=".2", edgecolor=".2")
plt.title('Bar Plot')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd # Import pandas

# Assuming the Titanic dataset is in a CSV file named 'titanic.csv'
titanic = sns.load_dataset('titanic') # Load the dataset

sns.boxplot(x= "sex",
            y="fare",
            data= titanic)
plt.show()
```

```python
sns.boxplot(x= "survived",
            y="age", showmeans=True,
            data= titanic)
plt.show()
```

```python
sns.boxplot(x= "survived",
            y="age", showmeans=True,
            meanprops= {"marker": "+",
                        "markersize":"12",
                        "markeredgecolor": "red"},
            data= titanic)
plt.show()

```

```python
sns.boxplot(x= "survived",
            y="age", showmeans=True,
            meanprops= {"marker": "*",
                        "markersize":"12",
                        "markeredgecolor": "red"},
            data= titanic)
plt.xlabel("How many survived", size ="12")
plt.ylabel("Age(years)", size = "12")
plt.title("Boxplot", size="15", weight="bold")
plt.show()
```

```python
# load dateset
dot= sns.load_dataset("dots")
dot
```

```python
sns.set_theme(style="ticks")
#defining a color pelette
p = sns.color_palette("Paired", 7)

#plot lineplot

sns.relplot(x="time", y="firing_rate", hue="coherence", size="choice",
            col="align", kind="line", size_order=["T1", "T2"], palette=p,
            height =5, aspect = .75, facet_kws=dict(sharex=False),
            data=dot)
```

```python
sns.set_theme(style="ticks", palette="pastel")

# Nested boxplot
sns.boxplot(x="day", y="total_bill",
            hue="smoker", palette=["m", "g"],
            data=tips)
sns.despine(offset=10, trim=True)
```

```python
sns.set_theme(style="whitegrid")

#nested violinplot
sns.violinplot(x="day", y="total_bill", hue="smoker",
               split=True, inner="quart", linewidth=1,
               palette={"Yes": "b", "No": ".85"},
               data = tips)

sns.despine(left=True)
```

```python
sns.set_theme(style="whitegrid")

# Load the example diamonds dataset
diamonds = sns.load_dataset("diamonds")

#scatter plot
f, ax = plt.subplots(figsize=(6.5,6.5))
sns.despine(f, left= True, bottom=True)
clarity_ranking=['I1','SI2','SI1', 'VS2','VS1','VVS2','VVS1','IF']
sns.scatterplot(x='carat',y='price',
                hue='clarity', size='depth',
                palette='ch:r=-.2,d=.3_r',
                hue_order=clarity_ranking,
                sizes=(1,8), linewidth=0,
                data=diamonds, ax=ax)
```

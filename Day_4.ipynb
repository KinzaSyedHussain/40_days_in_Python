# **Data** **Visualization**

```python
#import libraies
import seaborn as sns
import numpy
import scipy
import pandas
import matplotlib.pyplot as plt
```

```python
# load dataset
df = seaborn.load_dataset("iris")
df
```

```python
#draw a line plot
sns.lineplot(x='sepal_length',y='sepal_width',data=df)
plt.title('Line Plot')
plt.show()
```

```python
# Adding limits
sns.lineplot(x='sepal_length',y='sepal_width',data=df)
plt.title('Line Plot')
plt.xlim(4,8)
plt.ylim(2,4)
plt.show()
```

```python
#change figure
plt.figure(figsize=(10,5))

sns.lineplot(x='sepal_length',y='sepal_width',data=df)
plt.title('Line Plot')
plt.show()



```

```python
#setting style
sns.set_style("dark")

sns.lineplot(x='sepal_length',y='sepal_width',data=df)
plt.title('Line Plot')
plt.show()
```

```python
#barplot
sns.barplot(x='species',y='petal_length',data=df)
plt.title('Bar Plot')
plt.show()
```

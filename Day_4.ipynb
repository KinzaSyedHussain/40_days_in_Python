# **Data** **Visualization**

```python
#import libraies
import seaborn as sns
import numpy
import scipy
import pandas
import matplotlib.pyplot as plt
```

```python
# load dataset
df = seaborn.load_dataset("iris")
df
```

```python
#draw a line plot
sns.lineplot(x='sepal_length',y='sepal_width',data=df)
plt.title('Line Plot')
plt.show()
```

```python
# Adding limits
sns.lineplot(x='sepal_length',y='sepal_width',data=df)
plt.title('Line Plot')
plt.xlim(4,8)
plt.ylim(2,4)
plt.show()
```

```python
#change figure
plt.figure(figsize=(10,5))

sns.lineplot(x='sepal_length',y='sepal_width',data=df)
plt.title('Line Plot')
plt.show()



```

```python
#setting style
sns.set_style("dark")

sns.lineplot(x='sepal_length',y='sepal_width',data=df)
plt.title('Line Plot')
plt.show()
```

```python
#barplot
sns.barplot(x='species',y='petal_length',data=df)
plt.title('Bar Plot')
plt.show()
```

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dateset
data= sns.load_dataset("titanic")
data
```

```python
# draw a line plot
sns.lineplot(x='age',y='fare',data=data)
plt.title('Line Plot')
plt.show()
```

```python
#draw a barplot
sns.barplot(x='who',y='alone', hue = "sex", data=data)
plt.title('Bar Plot')
plt.show()
```

```python
#order
sns.barplot(x='who',y='alone', hue = "sex", data=data, order=['child','woman','man'])
plt.title('Bar Plot')
plt.show()
```

```python
# changing the color
sns.barplot(x='who',y='alone', hue = "sex", data=data, order=['child','woman','man'], color = 'red')
plt.title('Bar Plot')
plt.show()

sns.barplot(x='who',y='alone', hue = "sex", data=data, order=['child','woman','man'], color = 'pink')
plt.title('Bar Plot')
plt.show()
```

```python
#using colour pallete of seabor
sns.barplot(x='who',y='alone', hue = "sex", data=data, order=['child','woman','man'], ci = None, palette='bright' )
plt.title('Bar Plot')
plt.show()

sns.barplot(x='who',y='alone', hue = "sex", data=data, order=['child','woman','man'], ci = None, palette='pastel' )
plt.title('Bar Plot')
plt.show()
```
